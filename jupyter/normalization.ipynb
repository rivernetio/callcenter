```python
from sklearn.preprocessing import scale
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:.0f}'.format
pd.set_option('display.max_columns',100)
df=pd.read_csv('20180105_result.csv',encoding='gbk')
# 列出前五行
df.head()
```

```python
data = df[["MEAN_INTERVAL_CALL","SD_INTERVAL_CALL","NUM_CALL","MEAN_DURATION",
 "MOST_DURATION","MOST_DURATION_NUM","SD_DURATION","TOTAL_DURATION","TRK_NUM"]]
data = scale(data)

ndf = pd.DataFrame(data,columns=["MEAN_INTERVAL_CALL","SD_INTERVAL_CALL","NUM_CALL","MEAN_DURATION",
 "MOST_DURATION","MOST_DURATION_NUM","SD_DURATION","TOTAL_DURATION","TRK_NUM"],dtype='float64',)

ndf["CALLING_NUM"] = df["CALLING_NUM"]

ndf.to_csv('scale.csv',index=False)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3).fit(data)
```

```python
df["scaled_cluster"] = km.labels_
df.sort_values("scaled_cluster")
```

```python
df.groupby("scaled_cluster").mean()
```
